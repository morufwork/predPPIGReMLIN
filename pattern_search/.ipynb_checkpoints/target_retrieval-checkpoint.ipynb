```python
import os
import re
import pandas as pd

# Create an empty list to store DataFrame for each file
dataframes = []

# Specify the folder containing the files
folder_path = "pdbfiles"


# Loop through all files in the folder
lists = []
for filename in os.listdir(folder_path):
    if filename.endswith(".ent"):
        file_path = os.path.join(folder_path, filename)
        
        list = []
        # Open the file and read its contents
        list.append(filename)
        with open(file_path, 'r') as file:
            lines = file.readlines()
            #print(pdb_content)
            for i in range(0, len(lines)): 
                if "COMPND" in lines[i] and "MOLECULE" in lines[i]:
                    if "SPIKE" in lines[i]:
                        #print(lines[i])
                        #print(lines[i+1])
                        
                        molecom1 = lines[i]
                        chaincom1 = lines[i+1]
                        
                        mole1 = molecom1[:-1].rstrip()
                        mole1 = mole1.split(":")
                        molecule1 = mole1[1][:-1].rstrip()
                        
                        chain1 = chaincom1[:-1].rstrip()
                        chain1 = chain1.split(":")
                        chain1 =chain1[1][:-1].rstrip()

                    else:
                        #print(lines[i])
                        #print(lines[i+1])
                        
                        molecom2 = lines[i]
                        chaincom2 = lines[i+1]
                        
                        mole2 = molecom2[:-1].rstrip()
                        mole2 = mole2.split(":")
                        molecule2 = mole2[1][:-1].rstrip()
                        
                        chain2 = chaincom2[:-1].rstrip()
                        chain2 = chain2.split(":")
                        chain2 =chain2[1][:-1].rstrip()
        # Create a DataFrame for the current file
        data = {
            'File': [filename],
            'Molecule1': [molecule1],
            'Chain1': [chain1],
            'Molecule2': [molecule2],
            'Chain2': [chain2]
        }
     
        
        df = pd.DataFrame(data)
        dataframes.append(df)
       
    
# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(dataframes, ignore_index=True)

# Save the final DataFrame to a CSV file
output_file_path = "final_output.csv"
final_df.to_csv(output_file_path, index=False)

print("CSV file saved successfully:", output_file_path)                
```
