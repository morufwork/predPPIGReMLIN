```python
import networkx as nx

# Define the adjacency lists for the two bipartite graphs
# Replace these with your own adjacency lists
adj_list1 = {
    'A': ['a', 'b'],
    'B': ['c', 'd'],
}

adj_list2 = {
    'X': ['u', 'v'],
    'Y': ['w', 'x'],
}

# Create bipartite graphs from adjacency lists
G1 = nx.Graph()
G1.add_nodes_from(adj_list1.keys(), bipartite=0)
G1.add_nodes_from([node for sublist in adj_list1.values() for node in sublist], bipartite=1)
G1.add_edges_from((node1, node2) for node1, neighbors in adj_list1.items() for node2 in neighbors)

G2 = nx.Graph()
G2.add_nodes_from(adj_list2.keys(), bipartite=0)
G2.add_nodes_from([node for sublist in adj_list2.values() for node in sublist], bipartite=1)
G2.add_edges_from((node1, node2) for node1, neighbors in adj_list2.items() for node2 in neighbors)

# Check for isomorphism
isomorphism = nx.isomorphism.GraphMatcher(G1, G2)

if isomorphism.is_isomorphic():
    print("The two bipartite graphs are isomorphic.")
    # Print isomorphic nodes
    for node_pair in isomorphism.mapping.items():
        print(f"Corresponding nodes in G1 and G2: {node_pair}")
else:
    print("The two bipartite graphs are not isomorphic.")
```

```python
from networkx.algorithms import isomorphism

GM = isomorphism.GraphMatcher(Q, T)

if GM.subgraph_is_isomorphic():
    print("Subgraph is isomorphic to the larger graph.")
    print("Isomorphisms:")
    
    # Create a dictionary to store mappings based on chains
    chain_mappings = {}
    # Iterate over the mappings
    for subgraph_node, graph_node in GM.mapping.items():
        # Get the chain attribute of the corresponding nodes in both graphs
        subgraph_chain = Q.nodes[subgraph_node].get('chain')
        graph_chain = T.nodes[graph_node].get('chain')
        
        # Add the mapping to the corresponding chain in the dictionary
        if subgraph_chain in chain_mappings:
            chain_mappings[subgraph_chain].append((subgraph_node, graph_node))
        else:
            chain_mappings[subgraph_chain] = [(subgraph_node, graph_node)]

    # Print the mappings segmented by chain
    for chain, mappings in chain_mappings.items():
        print(f"Mappings for chain {chain}:")
        for subgraph_node, graph_node in mappings:
            print(f"Corresponding nodes in Q and T: {subgraph_node} and {graph_node}")
            
            
   
```

```python
from networkx.algorithms import isomorphism

GM = isomorphism.GraphMatcher(Q, T)

BQA = []
BTA = []
BQB = []
BTB = []

if GM.subgraph_is_isomorphic():
    print("Subgraph is isomorphic to the larger graph.")
    print("Isomorphisms:")
    
    # Create a dictionary to store mappings based on chains
    chain_mappings = {}
    QA = []
    TA = []
    QB = []
    TB = []
    # Iterate over the mappings
    for subgraph_node, graph_node in GM.mapping.items():
        # Get the chain attribute of the corresponding nodes in both graphs
        subgraph_chain = Q.nodes[subgraph_node].get('chain')
        graph_chain = T.nodes[graph_node].get('chain')
        
        # Add the mapping to the corresponding chain in the dictionary
        if subgraph_chain in chain_mappings:
            chain_mappings[subgraph_chain].append((subgraph_node, graph_node))
        else:
            chain_mappings[subgraph_chain] = [(subgraph_node, graph_node)]

    # Print the mappings segmented by chain
    for chain, mappings in chain_mappings.items():
        if chain == "A":
            print(f"Mappings for chain {chain}:")
            for subgraph_node, graph_node in mappings:
                print(f"Corresponding nodes in Q and T: {subgraph_node} and {graph_node}")
                QA.append(subgraph_node)
                TA.append(graph_node) 
                
        else:
            print(f"Mappings for chain {chain}:")
            for subgraph_node, graph_node in mappings:
                print(f"Corresponding nodes in Q and T: {subgraph_node} and {graph_node}")
                QB.append(subgraph_node)
                TB.append(graph_node) 
            
    BQA.append(QA)
    BTA.append(TA)
    BQB.append(QB)
    BTB.append(TB)
            

```
