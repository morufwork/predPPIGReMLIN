```python
import networkx as nx

# Create the larger graph with node and edge labels
G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4)])
nx.set_node_attributes(G, {0: "A", 1: "B", 2: "C", 3: "D", 4: "E"}, "label")
nx.set_edge_attributes(G, {(0, 1): "X", (0, 2): "Y", (1, 2): "Z", (1, 3): "W", (2, 3): "V", (3, 4): "U"}, "label")

# Create the subgraph with node and edge labels
subgraph = nx.Graph()
subgraph.add_edges_from([(0, 1), (0, 2)])
nx.set_node_attributes(subgraph, {0: "A", 1: "B", 2: "C"}, "label")
nx.set_edge_attributes(subgraph, {(0, 1): "X"}, "label")

# Check for subgraph isomorphism based on structure, node labels, and edge labels
def is_subgraph_isomorphic(larger_graph, subgraph):
    for node in subgraph.nodes():
        if node not in larger_graph.nodes():
            return False
    for edge in subgraph.edges():
        if edge not in larger_graph.edges():
            return False
    for node in subgraph.nodes():
        if subgraph.nodes[node].get("label") != larger_graph.nodes[node].get("label"):
            return False
    for edge in subgraph.edges():
        if subgraph.edges[edge].get("label") != larger_graph.edges[edge].get("label"):
            return False
    return True

is_isomorphic = is_subgraph_isomorphic(G, subgraph)

if is_isomorphic:
    print("Subgraph is isomorphic to the larger graph.")
    print("subgraph:", subgraph)
    print("G:", G)
else:
    print("Subgraph is not isomorphic to the larger graph.")
    print("subgraph:", subgraph)
    print("G:", G)
```

```python
import networkx as nx
from networkx.algorithms import isomorphism

# Create the larger graph G with node and edge labels
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (4, 5)])
nx.set_node_attributes(G, {1: "A", 2: "B", 3: "C", 4: "D", 5: "E"}, "label")
nx.set_edge_attributes(G, {(1, 2): "X", (1, 3): "Y", (2, 3): "Z", (2, 4): "W", (3, 4): "V", (4, 5): "U"}, "label")

# Create the subgraph subgraph with node and edge labels
subgraph = nx.Graph()
subgraph.add_edges_from([(1, 2), (2, 3)])
nx.set_node_attributes(subgraph, {1: "A", 2: "B", 3: "C"}, "label")
nx.set_edge_attributes(subgraph, {(1, 2): "X"}, "label")

# Check for subgraph isomorphism based on structure, node labels, and edge labels
GM = isomorphism.GraphMatcher(G, subgraph, node_match=isomorphism.categorical_node_match("label", "label"), edge_match=isomorphism.categorical_edge_match("label", "label"))

if GM.subgraph_is_isomorphic():
    print("Subgraph is isomorphic to the larger graph.")
else:
    print("Subgraph is not isomorphic to the larger graph.")
```

```python
import networkx as nx
from networkx.algorithms import isomorphism

# Create the larger graph G with node and edge labels
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (4, 5)])
nx.set_node_attributes(G, {1: "A", 2: "B", 3: "C", 4: "D", 5: "E"}, "label")
nx.set_edge_attributes(G, {(1, 2): "X", (1, 3): "Y", (2, 3): "Z", (2, 4): "W", (3, 4): "V", (4, 5): "U"}, "label")

# Create the subgraph subgraph with node and edge labels
subgraph = nx.Graph()
subgraph.add_edges_from([(1, 2), (1, 3), (2, 3)])
nx.set_node_attributes(subgraph, {1: "A", 2: "B", 3: "C"}, "label")
nx.set_edge_attributes(subgraph, {(1, 2): "X", (1, 3): "Y", (2, 3): "Z"}, "label")


# Check for subgraph isomorphism based on structure, node labels, and edge labels
GM = isomorphism.GraphMatcher(G, subgraph, node_match=isomorphism.categorical_node_match("label", "label"), edge_match=isomorphism.categorical_edge_match("label", "label"))

if GM.subgraph_is_isomorphic():
    print("Subgraph is isomorphic to the larger graph.")
    print("Isomorphisms:")
    for isomorphism in GM.subgraph_isomorphisms_iter():
        print("Isomorphic area:")
        # Extract the nodes and edges corresponding to the subgraph in G
        subgraph_nodes = set(subgraph.nodes())
        isomorphic_nodes = set(isomorphism.keys())
        
        #isomorphic_edges = set((node1, node2) for node1, node2 in isomorphism.items() if G.has_edge(node1, node2))
        
        # Find the isomorphic edges by checking if edges in the subgraph exist in G
        isomorphic_edges = {(node1, node2) for node1, node2 in subgraph.edges() if (node1, node2) in G.edges()}


        # Print the nodes and edges in the isomorphic area
        print("Nodes:")
        print(isomorphic_nodes)
        print("Edges:")
        print(isomorphic_edges)
        print("------------------")
else:
    print("Subgraph is not isomorphic to the larger graph.")
```

```python
import networkx as nx
from networkx.algorithms import isomorphism

# Create the larger graph G with node and edge labels
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (4, 5)])
nx.set_node_attributes(G, {1: "A", 2: "B", 3: "C", 4: "D", 5: "E"}, "label")
nx.set_edge_attributes(G, {(1, 2): "X", (1, 3): "Y", (2, 3): "Z", (2, 4): "W", (3, 4): "V", (4, 5): "U"}, "label")

# Create the subgraph subgraph with node and edge labels
subgraph = nx.Graph()
subgraph.add_edges_from([(1, 2), (1, 3), (2, 3)])
nx.set_node_attributes(subgraph, {1: "A", 2: "B", 3: "C"}, "label")
nx.set_edge_attributes(subgraph, {(1, 2): "X", (1, 3): "Y", (2, 3): "Z"}, "label")

# Convert dictionaries to NetworkX graphs
data_graph = nx.Graph(G)
query_graph = nx.Graph(subgraph)

# Check for subgraph isomorphism based on structure, node labels, and edge labels
GM = isomorphism.GraphMatcher(data_graph, query_graph, node_match=isomorphism.categorical_node_match("label", "label"), edge_match=isomorphism.categorical_edge_match("label", "label"))

if GM.subgraph_is_isomorphic():
    print("Subgraph is isomorphic to the larger graph.")
    print("Isomorphisms:")
    for isomorphism in GM.subgraph_isomorphisms_iter():
        print("Isomorphic area:")
        # Extract the nodes corresponding to the subgraph in G
        subgraph_nodes = set(query_graph.nodes())
        isomorphic_nodes = set(isomorphism.keys())
        
        # Find the isomorphic edges by checking if edges in the subgraph exist in G
        isomorphic_edges = {(node1, node2) for node1, node2 in query_graph.edges() if (node1, node2) in data_graph.edges()}

        # Print the nodes and edges in the isomorphic area
        print("Nodes:")
        print(isomorphic_nodes)
        print("Edges:")
        print(isomorphic_edges)
        print("------------------")
else:
    print("Subgraph is not isomorphic to the larger graph.")
```
