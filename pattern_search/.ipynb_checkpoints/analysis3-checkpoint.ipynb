```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("haddock3.csv")

# Convert to long format for one metric (e.g., HADDOCK score)
df_long = df.melt(
    id_vars="Sample",
    value_vars=["Docking_HADDOCK", "Redocking_HADDOCK", "Null_HADDOCK"],
    var_name="Condition",
    value_name="HADDOCK_score"
)

# Clean up condition names
df_long["Condition"] = df_long["Condition"].str.replace("_HADDOCK", "")
print(df_long.head())
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df_long, x="Condition", y="HADDOCK_score", palette="Set2")
sns.swarmplot(data=df_long, x="Condition", y="HADDOCK_score", color="black", alpha=0.6)
plt.title("Distribution of HADDOCK Scores Across Docking Conditions")
plt.show()
```

```python
sns.violinplot(data=df_long, x="Condition", y="HADDOCK_score", inner="quartile", palette="muted")
plt.title("HADDOCK Score Density per Condition")
plt.show()
```

```python
sns.pairplot(df, hue='Type', vars=['Docking_HADDOCK', 'Docking_BSA', 'Docking_Evdw', 'Docking_Eelec', 'Docking_Edesolv'])
plt.show()

corr = df[['Docking_HADDOCK','Docking_BSA','Docking_Evdw','Docking_Eelec','Docking_Edesolv']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between HADDOCK Metrics')
plt.show()
```

```python
from scipy.stats import shapiro

for col in ["Docking_HADDOCK", "Redocking_HADDOCK", "Null_HADDOCK"]:
    stat, p = shapiro(df[col])
    print(f"{col}: p={p:.4f}")
```

```python
import pingouin as pg  # install via pip install pingouin

df_anova = df_long.rename(columns={"Condition": "within", "HADDOCK_score": "DV"})
anova = pg.rm_anova(dv="DV", within="within", subject="Sample", data=df_anova, detailed=True)
print(anova)
```

```python
from scipy.stats import friedmanchisquare

stat, p = friedmanchisquare(df["Docking_HADDOCK"], df["Redocking_HADDOCK"], df["Null_HADDOCK"])
print(f"Friedman Test HADDOCK Score: ={stat:.3f}, p={p:.4f}")
```

```python
from scipy.stats import wilcoxon

pairs = [("Docking_HADDOCK", "Redocking_HADDOCK"),
         ("Docking_HADDOCK", "Null_HADDOCK"),
         ("Redocking_HADDOCK", "Null_HADDOCK")]

for a, b in pairs:
    stat, p = wilcoxon(df[a], df[b])
    print(f"{a} vs {b}: p={p:.4f}")
```

```python
plt.figure(figsize=(7,5))
sns.lineplot(data=df_long, x="Condition", y="HADDOCK_score", hue="Sample", marker="o", legend=False, alpha=0.5)
plt.title("Per-Sample Change in HADDOCK Score Across Conditions")
plt.show()
```

```python
summary = df_long.groupby("Condition")["HADDOCK_score"].agg(['mean', 'std', 'min', 'max'])
summary.to_csv("haddock_summary_stats.csv")
print(summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Input Data ===
data = {
    "Metric": ["HADDOCK", "BSA", "Evdw", "Eelec", "Edesolv"],
    "Docking_vs_Redocking_p": [0.0000, 0.0260, 0.0982, 0.0011, 0.0274],
    "Docking_vs_Null_p": [0.0000, 0.0000, 0.0005, 0.0001, 0.1821],
    "Redocking_vs_Null_p": [0.0010, 0.0000, 0.0246, 0.0829, 0.4214],
}

df = pd.DataFrame(data)

# === Significance labeling function ===
def significance(p):
    if p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return "ns"

# Add significance annotations
df["Docking_vs_Redocking_sig"] = df["Docking_vs_Redocking_p"].apply(significance)
df["Docking_vs_Null_sig"] = df["Docking_vs_Null_p"].apply(significance)
df["Redocking_vs_Null_sig"] = df["Redocking_vs_Null_p"].apply(significance)

# === Save to Excel ===
excel_path = "HADDOCK_significance_summary.xlsx"
df.to_excel(excel_path, index=False)
print(f" Excel summary saved as: {excel_path}")

# === Visualization ===
fig, ax = plt.subplots(figsize=(9, 5))
bar_positions = range(len(df))
bar_width = 0.25

# Create grouped bar plot
ax.bar([p - bar_width for p in bar_positions], df["Docking_vs_Redocking_p"], width=bar_width, label="Docking vs Redocking")
ax.bar(bar_positions, df["Docking_vs_Null_p"], width=bar_width, label="Docking vs Null")
ax.bar([p + bar_width for p in bar_positions], df["Redocking_vs_Null_p"], width=bar_width, label="Redocking vs Null")

# Add significance stars above each bar
for i, metric in enumerate(df["Metric"]):
    for j, col in enumerate(["Docking_vs_Redocking_p", "Docking_vs_Null_p", "Redocking_vs_Null_p"]):
        sig = significance(df.loc[i, col])
        xpos = i + (j - 1) * bar_width
        ax.text(xpos, df.loc[i, col] + 0.02, sig, ha="center", va="bottom", fontsize=12, fontweight="bold")

# Add reference line and labels
ax.axhline(0.05, color="red", linestyle="--", label="p = 0.05 threshold")
ax.set_xticks(bar_positions)
ax.set_xticklabels(df["Metric"], fontsize=11)
ax.set_ylabel("p-value")
ax.set_title("Pairwise Wilcoxon Test Results (Docking, Redocking, Null)")
ax.legend()
plt.tight_layout()

# Save figure
plot_path = "HADDOCK_significance_plot.png"
plt.savefig(plot_path, dpi=300)
plt.show()

print(f" Plot saved as: {plot_path}")
```
