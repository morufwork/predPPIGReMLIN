```python
import matplotlib.pyplot as plt

# X-axis labels
thresholds = ["NA", "0.3", "0.4", "0.5", "0.6"]
x_positions = range(len(thresholds))

# --- AUC values ---
auc_blue = [0.9996, 0.99758, 0.99771, 0.99679, 0.98985]
auc_green = [0.87, 0.71, 0.70, 0.69, 0.69]
auc_purple = [0.81, 0.65, 0.64, 0.62, 0.61]
auc_skyblue = [0.80, 0.63, 0.62, 0.61, 0.60]
auc_orange = [0.84, 0.58, 0.56, 0.55, 0.54]

# --- AUPR values ---
aupr_blue = [0.99997, 0.999876, 0.999874, 0.999753, 0.997686]
aupr_green = [0.65, 0.38, 0.36, 0.38, 0.36]
aupr_purple = [0.60, 0.29, 0.28, 0.29, 0.28]
aupr_skyblue = [0.52, 0.26, 0.25, 0.25, 0.24]
aupr_orange = [0.58, 0.18, 0.18, 0.19, 0.18]

# --- Create subplots side by side ---
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharex=True)

# ===== Plot AUC =====
axes[0].plot(x_positions, auc_blue, color='blue', marker='o', label='predPPI')
axes[0].plot(x_positions, auc_green, color='green', marker='o', linestyle='-', label='CAMP')
axes[0].plot(x_positions, auc_purple, color='purple', marker='o', linestyle='--', label='DeepDTA')
axes[0].plot(x_positions, auc_skyblue, color='deepskyblue', marker='o', linestyle='--', label='PIPR')
axes[0].plot(x_positions, auc_orange, color='orange', marker='o', linestyle=':', label='NRLMF')

axes[0].set_xticks(list(x_positions))
axes[0].set_xticklabels(thresholds)
axes[0].set_xlabel("Clustering Threshold")
axes[0].set_ylabel("AUC")
axes[0].set_title("AUC vs Clustering Threshold")
axes[0].set_ylim(0.5, 1.02)
axes[0].grid(True)
axes[0].legend(loc="lower left")

# ===== Plot AUPR =====
axes[1].plot(x_positions, aupr_blue, color='blue', marker='o', label='predPPI')
axes[1].plot(x_positions, aupr_green, color='green', marker='o', linestyle='-', label='CAMP')
axes[1].plot(x_positions, aupr_purple, color='purple', marker='o', linestyle='--', label='DeepDTA')
axes[1].plot(x_positions, aupr_skyblue, color='deepskyblue', marker='o', linestyle='--', label='PIPR')
axes[1].plot(x_positions, aupr_orange, color='orange', marker='o', linestyle=':', label='NRLMF')

axes[1].set_xticks(list(x_positions))
axes[1].set_xticklabels(thresholds)
axes[1].set_xlabel("Clustering Threshold")
axes[1].set_ylabel("AUPR")
axes[1].set_title("AUPR vs Clustering Threshold")
axes[1].set_ylim(0.15, 1.02)
axes[1].grid(True)
axes[1].legend(loc="lower left")

plt.tight_layout()
plt.show()
```
